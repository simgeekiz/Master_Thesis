```python
events = {}
for sent in art1.doc.sentences():
    print(sent)
    
    print('\nENTITIES\n----------\ncls, \tent')
    for ent in sent.select(folia.Entity):
        print(ent.cls, '\t', ent, end='')
        
        coms = [c for c in ent.select(folia.Comment)]
        if coms:
            for c in coms:
                print('\t', c.json()['value'])
                
                val = c.json()['value']
 
        else:
            val = 'Event 1'
                
            print('\t', 'Event 1')
            
        if val not in events.keys():
            events[val] = []

        if sent.text() not in events[val]:
            events[val].append(sent.text())
    
    print('##################\n')
```
